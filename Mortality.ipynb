```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
d = pd.read_csv('https://www.mortality.org/Public/STMF/Outputs/stmf.csv', header = 1, dtype = {'CountryCode' : "string", 'Year': int, 'Week' : int, \
     'Sex': "string", 'D0_14': float, 'D15_64': float, 'D65_74': float, 'D75_84': float, 'D85p': float,  \
     'DTotal': float, 'R0_14': float, 'R15_64': float, 'R65_74': float, 'R75_84': float, 'R85p': float,  \
     'RTotal': float, 'Split': int,   'SplitSex': int, 'Forecast': int })
```

```python
USA_data_both = d[ (d['CountryCode']=='USA') & (d['Sex']=='b') ]  
```

```python
fig, ax = plt.subplots(figsize=(12,8))
for i in range(2011,2021):
    USA_data_both[USA_data_both['Year'] == i].plot( x='Week', y='RTotal', label=str(i),ax = ax);
```
